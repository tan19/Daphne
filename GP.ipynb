```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from ml.models.gaussian_process_regression import GaussianProcessRegression
```

```python
import numpy as np

X = np.linspace(start=0, stop=10, num=100).reshape(-1, 1)
y_true = np.squeeze(X * np.sin(X))
y = y_true + 0 * np.random.randn(len(y_true))

rng = np.random.RandomState(1)
training_indices = rng.choice(np.arange(y.size), size=10, replace=False)
X_train, y_train = X[training_indices], y[training_indices]

GPR = GaussianProcessRegression(X_train, y_train, X)

plt.plot(X, y_true, label=r"$f(x) = x \sin(x)$", linestyle="dotted")
plt.scatter(X_train, y_train, label="Observations")

plt.plot(X, GPR.conditional_mean, label="Mean prediction")
```
